```python
print("Here you have our quiz !")
print("--------------------")

print("Question 1")
question_1=input("What is the capital of France?")
while question_1.lower()!="paris":
    print ("Your answer :",question_1)
    print ("Too bad! that is not the correct answer, Try again") 
    question_1=input("What is the capital of France?")
else:
    print ("Your answer :",question_1) 
    print ("Good job! This is the right answer")

print("Question 2")
question_2=input("What is the last name of the French President?")
while question_2.lower()!="macron":
    print ("Your answer :",question_2)
    print ("Too bad! that is not the correct answer, Try again") 
    question_2=input("What is the last name of the French President?")
else: 
    print ("Your answer :",question_2)
    print ("Good job! This is the right answer")

print("Question 3")
question_3=input("What is the name of the second largest city in France ?")
while question_3.lower()!="marseille":
    print ("Your answer :",question_3)
    print ("Too bad! that is not the correct answer, Try again") 
    question_3=input("What is the name of the second largest city in France ?")
else: 
    print ("Your answer :",question_3)
    print ("Good job! This is the right answer")

print("Well done you are the quiz winner")
```

```python
print("Here you have our quiz !")
print("You have 3 lives.")
print("--------------------")

lives=3
print("Question 1")
question_1=input("What is the capital of France?")

while question_1.lower()!="paris":
    print ("Your answer :",question_1)
    lives-=1
    print("Sorry, you have {} chances left".format(lives))
    if lives==0:
        print("You lose")
        break
    question_1=input("What is the capital of France?")

else:
    print ("Your answer :",question_1) 
    print ("Good job! This is the right answer")

print("--------------------")

print("Question 2")
question_2=input("What is the last name of the French President?")
while question_2.lower()!="macron":
    print ("Your answer :",question_2)
    lives-=1
    print("Sorry, you have {} chances left".format(lives))
    if lives==0:
        print("You lose")
        break
    question_2=input("What is the last name of the French President?")

else:
    print ("Your answer :",question_2) 
    print ("Good job! This is the right answer")

print("--------------------")

print("Question 3")
question_3=input("What is the name of the second largest city in France ?")
while question_3.lower()!="marseille":
    print ("Your answer :",question_3)
    lives-=1
    print("Sorry, you have {} chances left".format(lives))
    if lives==0:
        print("You lose")
        break
    question_3=input("What is the name of the second largest city in France ?")

else:
    print ("Your answer :",question_3) 
    print ("Good job! This is the right answer")

print("Well done you are the quiz winner")
```
